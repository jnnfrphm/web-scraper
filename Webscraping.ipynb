```python
import pandas as pd
import re
import time


from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
url = "https://search.vivli.org/?search=a%20NOT%20a"
```

```python
wd = webdriver.Chrome('/Users/jenniferpham/Documents/GitHub/entry-level-jobs/chromedriver')
wd.get(url)
```

```python
elements = wd.find_elements_by_class_name('searchResult_viewDetails')
print(len(elements))
count = 0
for x in range(0, len(elements)):
    if elements[x].is_displayed():
        elements[x].click()
        count += 1
```

```python
phases = []
diseases = []
treatments = []
sample_sizes = []
IDs = []
links = []

print(count)
for i in range(1, count+1):
    wd.switch_to.window(wd.window_handles[i])
    
    phase = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[1]/div[1]/div/div/input""")
    phases.append(phase.get_attribute('value'))
    
    disease = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[1]/div[2]/div/div/input""")
    diseases.append(disease.get_attribute('value'))

    treatment = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[1]/div[3]/div[1]/div/div/textarea""")
    treatments.append(treatment.get_attribute('placeholder'))

    sample_size = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[2]/div[4]/div/div/input""")
    sample_sizes.append(sample_size.get_attribute('value'))

    wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[2]/div/div/div/div[2]""").click()
    time.sleep(3)
    
    try:
        ID = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[1]/div[5]/div[2]/div/div/input""")
        IDs.append(ID.get_attribute('value'))
    except:
        IDs.append(ID.get_attribute('N/A'))
        
    link = wd.find_element_by_xpath("""//*[@id="modal-blur-container"]/div[2]/div[3]/div[1]/div[5]/div[3]/div/div/div[2]/a""")
    links.append(link.get_attribute('href'))
```

```python
print(len(phases))
print(len(diseases))
print(len(treatments))
print(len(sample_sizes))
print(len(IDs))
print(len(links))
```

```python
data = pd.DataFrame({'ID': IDs,
'Disease': diseases,
'Treatment': treatments,                        
'Sample Size': sample_sizes,
'Phase': phases,
'Link': links,
})

print(data.info())
data
```

```python
data.to_csv('data5.csv', index=0)
```
